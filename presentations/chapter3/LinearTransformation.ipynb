```python
import sympy as sym
sym.init_printing()
```

```python
A = sym.Matrix([[4, -2, 5, -5, 0],
                [-9, 7, -8, 0, 0],
                [-6, 4, 5, 3, 0],
                [5, -3, 8, -4, 0]])
display(A.rref())
```
