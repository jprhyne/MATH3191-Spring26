````python
import numpy as np
````

````python
def proj(x, v):
    # Return a vector that is x projected onto v
    ret_vec = (np.dot(x,v) / np.dot(v,v)) * v
    return ret_vec
````

````python
def gramSchmidt( A ):
    """
    A is a 2d numpy array whose columns are the vectors (A[:,i] is column i)

    We return a matrix Q such that Q^T Q = I using Gram-Schmidt
    """
    (rows, columns) = A.shape
    Q = np.zeros((rows, columns))
    for i in range(columns):
        ui = A[i,:]
        for k in range(i):
            ui = ui + proj(A[i, :], Q[k,:])
        Q[:, i] = ui / np.linalg.norm(ui)
    return Q
````
